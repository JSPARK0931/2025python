```python
# 컬럼을 두개를 합쳐서 하나로 1차원-> 2차원
np.column_stack(([1,2,3],[4,5,6]))
```

```python
fish_data  = np.column_stack((fish_length, fish_weight))
print(fish_data)
print(fish_data.shape)
```

```python
fish_data[:5]
```

```python
# 1 로채움 :np.ones, 0으로 채움: np.zeros 
np.ones(35)
np.zeros(34)
```

```python
fish_target = np.concatenate((np.ones(35), np.zeros(14)))
fish_target
```

```python
fish_target.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(fish_data, fish_target, random_state=42)
```

```python
# KNeighborsClassifier() : 기본 갯수는 5개를 계산함
kn = KNeighborsClassifier()
kn.fit(X_train, y_train)
kn.score(X_test, y_test)
```

```python
# 빙어가 아닌데, 빙어로 나옴 ( 크기로 봐서는 도미일듯)
print(kn.predict([[25,150]]))
```

### train_scaled = (X_train - min) / std

```python
# 평균값
mean = np.mean(X_train,axis=0)
mean
```

```python
# 표준편차
std = np.std(X_train,axis=0)
std
```

```python
# 0을 기준으로 -1, 1사이 값으로 만들어줌
train_scaled = (X_train - mean) / std
train_scaled
```

```python
# 스케일 조정된 값으로 scatter 
# random 값 train_scaled 0번째 컬럼, train_scaled 1번째 컬럼 : blue
#plt.scatter(X_train[:,0], X_train[:,1])

new = ([25,150] - mean) /std

plt.scatter(train_scaled[:,0], train_scaled[:,1])
#plt.scatter(25,150,marker="^")
plt.scatter(new[0], new[1], marker="^")
plt.show()
```

```python

train_scaled.shape
```

```python
#스케일 조정값으로 fit
kn.fit(train_scaled, y_train)

test_scaled = (X_test - mean) / std

kn.score(test_scaled, y_test)
```

```python
# 조정된 값으로 predict
# new = ([25,150] - mean) /std
# 이전 scale 조정전은 0 빙어 , scale 조정후 1 도미로 나옴
print(kn.predict([new]))
print([new])
```

```python
distances,indexes = kn.kneighbors([new])
```
