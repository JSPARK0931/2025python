### 동물별 평균체온과 최대 체중구하기

```python
import numpy as np

# 체온, 심박수, 체중(kg) 
data = np.array([
    [38.5, 80, 35], #개
    [39.1, 90, 20], #고양이
    [36.8, 60, 400] #말
])

ani = np.array(["개", "고양이", "말"])

#체중, 심박수, 체중 평균값
# axis = 0 행, axis = 1 컬럼
avg_value = data.mean(axis=0)
print(avg_value)

#체중
weight = data[:,2]
print(weight)
#쵀대체중
max_weight = data[:,2].max()
print(max_weight)

max_ani = ani[data[:,2].argmax()]
print(max_ani)

print("동물별 평균 수치 :", avg_value)
print("가장무거운동물   :", max_ani, f"{max_weight} kg")
```

### 곤충 : 날개수, 다리수, 몸길이(cm)

```python
data = np.array([
    [2, 6, 1.2],
    [4, 6, 3.4],
    [0, 6, 0.8],
    [0, 8, 2.0]
])

insects = np.array(["나비", "잠자리", "개미", "거미"])

# 다리가 6개인 곤충만 추출
legs = data[data[:,1] == 6]
print(legs)

# 평균 몸길이 구하기
mean_length = data[:,2].mean()
print(mean_length)
print(f"곤충의 평균 몸길이 : ", round(mean_length,2), "cm")

# 날개가 없는 곤충
no_wings = insects[data[:,0] == 0]
#print(no_wings)
print(f"날개없는 곤충 :", no_wings)
```
