```python
st2 = 'abc'
for i in range(len(st2)):
    print(st2[i], end="-")

print("\n")
# 마지막 -를 안넣고 싶을때 
# true if 조건 else false 이용

#st2 = 'abcdef'
st2= ["a", "b", "c"]

for i in range(len(st2)):
    print(st2[i], end="" if i == len(st2)-1 else "-")
```

```python
#입력한문자를 소문자로 변경
lower_list = []

while True:
    s = input("영어단어를 입력 : (종료 :exit)")
    if s == "exit":
        break
    lower_list.append(s.lower())

for word in lower_list:
    print(word)
```
