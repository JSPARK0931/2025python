```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# numpy file 을 읽어옮
fruits = np.load("data/fruits_300.npy")
```

```python
# 3차원 data ( image )
fruits.shape
```

```python
# 0 : 검은색 , 255 : 밝은색
fruits[0,0,:]
```

```python
# 이미지출력
plt.imshow(fruits[0])
plt.show()
# 흑백출력
plt.imshow(fruits[0], cmap='gray')
plt.show()
# 흑백반전
plt.imshow(fruits[0], cmap='gray_r')
```

```python
# 이미지출력
plt.imshow(fruits[202])
plt.show()
```

```python
# matplotlib 이용 한행에 이미지 2개 - 보통 gray를 많이 사용함
fig, axs = plt.subplots(1,3)
axs[0].imshow(fruits[0], cmap = 'gray_r')
axs[1].imshow(fruits[100], cmap = 'gray_r')
axs[2].imshow(fruits[200], cmap = 'gray_r')
plt.show()

fig, axs = plt.subplots(2,3)
axs[0,0].imshow(fruits[0])
axs[0,1].imshow(fruits[100], cmap = 'gray')
axs[0,2].imshow(fruits[200], cmap = 'gray_r')
axs[1,0].imshow(fruits[50], cmap = 'gray_r')
axs[1,1].imshow(fruits[150], cmap = 'gray_r')
axs[1,2].imshow(fruits[250], cmap = 'gray_r')
plt.show()
```

```python
# fixel값 분석
# 100 X 100 이므로 10000줄 생성됨 3차원 => 2차원
apple = fruits[0:100].reshape(100, 100*100)  # 100,100,100
fineapple = fruits[100:200].reshape(-1,100*100)  # 100,100,100  : -1 을 나머지
banana = fruits[200:300].reshape(-1,100*100)
```

```python
print(apple.shape, fineapple.shape, banana.shape)
```

```python
apple.mean(axis=1)
```

```python
plt.hist(np.mean(apple,axis=1))
plt.hist(np.mean(fineapple,axis=1))
plt.hist(np.mean(banana,axis=1))
plt.legend(['apple','pineapple','banana'])
```

```python
# apple, pineapple 의 값이 중첩 : 알파는 투명도
plt.hist(np.mean(apple,axis=1 ), alpha=0.5)
plt.hist(np.mean(fineapple,axis=1), alpha=0.5)
plt.hist(np.mean(banana,axis=1), alpha=0.5)
plt.legend(['apple','pineapple','banana'])
```

```python
# 평균값 확인 3차원배열
fruits1 = np.array([
    [[1,2],
     [3,4]], # image 0
    [[2,2],[2,2]], # image 1
    [[4,4],[4,4]]  # image 2
])
```

```python
# axis = 1 가로
print(fruits1[0].mean(axis=1))
print(fruits1[1].mean(axis=1))
print(fruits1[2].mean(axis=1))
print("---------------------------------------")
# axis = 0 세로
print(fruits1[0].mean(axis=0))
print(fruits1[1].mean(axis=0))
print(fruits1[2].mean(axis=0))
```

```python
fig,axs = plt.subplots(1,3, figsize=(20,5))
axs[0].bar(range(10000), np.mean(apple, axis=0))
axs[1].bar(range(10000), np.mean(fineapple, axis=0))
axs[2].bar(range(10000), np.mean(banana, axis=0))
plt.show()
```
