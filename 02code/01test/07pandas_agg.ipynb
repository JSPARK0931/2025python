```python
import pandas as pd
```

```python
data = {
    "class": ["A", "A", "B", "B"],
    "name": ["철수", "영희", "민수", "지수"],
    "score": [80, 90, 70, 85]
}

#dataframe으로 변환
df = pd.DataFrame(data)
df
```

```python
#group by 객체생성
df.groupby("class")

for key, value in df.groupby("class"):
    print(f"key :{key} / value : {value}")
```

```python
for key, value in df.groupby("class"):
    print(key)
    print(value)
```

```python
df.groupby("class")["score"].mean()
```

```python
data1 = {
    '학급': ['A', 'A', 'B', 'B', 'C', 'C'],
    '이름': ['철수', '영희', '민수', '수진', '지훈', '혜진'],
    '수학': [90, 80, 70, 85, 60, 95],
    '영어': [85, 88, 75, 70, 95, 90]
}

df1 = pd.DataFrame(data1)
df1
```

```python
#학급별 수학,영어점수평균, 인덱스 추가
df1_mean = df1.groupby("학급")[["수학","영어"]].mean().reset_index()
```

```python
df1_mean
```

```python
df2_mean = df1.groupby("학급")[["수학","영어"]].mean()
df2_mean
```
