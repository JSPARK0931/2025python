### Dictionary

```python
s1 = [1,2,3]
d1 = {'kor':1 , 'eng':2, 'ja':3}

print(s1[0])
print(d1['kor'])
#print(d1.kor) - dictionary 사용불가
print(d1.get('kor'))

# 딕셔너리 항목 추가 d[키]=value
d1['ger'] = 4
print(d1)
```

```python
prices ={"아메리카노":2000, "라떼":3000, "녹차":2500}
```

```python

for key in prices:
    #print(key, end="" if key =="녹차" else "-")
    #print(key,end="-")
    #print(prices[key])
    print(f"{key}-{prices[key]}", end=",")
```

```python
# for in 문
for key, value in prices.items():
    print(f'{key}:{value}', end ="원 ")
```

```python
#pop으로 key:value 삭제
a = prices.pop("녹차")
print(a)
print(prices)
```

```python
del prices["라떼"]
print(prices)
```
