```python
movies = movies_metadata_small[['id', 'title', 'genres', 'keywords', 'overview', 'popularity', 'release_date']]
```

```python
movies.head()
```

```python
# 데이터 전처리 , genres, keywords, overview 를 섞어서 contents를 만듦 
# 참조 : bowmovie
```

```python
movies['str_genres'] = movies['genres'].fillna('[]') \
                    .apply(literal_eval) \
                    .apply(lambda x:sorted(i['name'] for i in x) if isinstance(x,list) else []) \
                    .apply(lambda x: ", ".join(x) if len(x) > 0 else None)
```

```python
movies.head()
```

```python
movies['str_keywords'] = movies['keywords'].fillna('[]') \
                    .apply(literal_eval) \
                    .apply(lambda x:sorted(i['name'] for i in x) if isinstance(x,list) else []) \
                    .apply(lambda x: ", ".join(x) if len(x) > 0 else None)
```

```python
movies['str_keywords']
```

```python
def get_genres_keywords_overview(row):
    return f"""
    genres:
    {row['str_genres']}

    keywords:
    {row['str_keywords']}

    overview:
    {row['overview']}

    """
```

```python
#axis = 0 : row , axis=1 : column
movies['genres_keywords_overview'] = movies.apply(get_genres_keywords_overview, axis=1)
```

```python
movies['genres_keywords_overview']
```

```python
movies['genres_keywords_overview'].iloc[0]
```

```python
#year
movies['release_date'] = pd.to_datetime(movies['release_date'], errors='coerce')
movies['year'] = movies['release_date'].dt.year
```

```python
#popularity_log 
movies['popularity'] = movies['popularity'].astype(float)
movies['popularity_log'] = np.log(movies['popularity'])
```

```python
# drop null data 결측치 제거후 index 재생성
movies = movies.dropna().reset_index(drop=True)
```
