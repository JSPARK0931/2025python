```python
# dictionary 데이터 프레이임으로 전환
# dic1 = {
#     'col1':[1,2,3,4],
#     'col2':[5,6,7,8]
# }
# n= pd.DataFrame(dic1)

n = pd.DataFrame({
    'col1':['1','2','3','4'],
    'col2':[5,6,7,8]
})
print(n)

```

```python
n.info()
n.dtypes
# 숫자 column만 통계자료 생성됨
n.describe()
```

```python
#데이터 type변경 (String -> int), (int -> float)
n_df = n.astype({'col1':int, 'col2':float})
n_df.dtypes
#컬럼을 숫자type으로 변경시 통계자료 생성됨
n_df.describe()
```

```python
#전체컬럼조회
display(n_df)
#컬럼값1개는 대괄호가 1개 필요
display(n_df['col1'])
#컬럼값2개 이상에는 대괄호가 2개 필요
display(n_df[['col1','col2']]) 
```

```python
# 순회해서 데이터 접근하기 : iteration rows : row 순회함수
for index, row in n_df.iterrows():
    print(row['col1'])
    
```

```python
# 새로운 컬럼을 추가해서 sum값 만들기
n_df
n_df['sum'] = n_df['col1'] + n_df['col2']
n_df
```

```python
# 8이상인sum값 구하기
n1_df = n_df[n_df['sum']>= 8]
```

```python
# index가 빠진상태로  들어간다
n_df[n_df['sum'] >= 8]
```

```python
# index 초기화 : 원래 인덱스는 남고, 새로운 초기화된 index 생성됨
n1_df.reset_index()
n1_sum = n1_df['sum']
print(type(n1_sum))
print(n1_sum.shape)
print(n1_sum)
```

```python
# index를 초기화하는 다른방법 : drop=True : 예전 index는 사라지고 새로운 인덱스 생김
n2_df = n_df[n_df['sum']>= 8].reset_index(drop=True)
```
